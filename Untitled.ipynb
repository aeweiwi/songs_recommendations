```python
ds = pd.read_csv('kaggle_visible_evaluation_triplets.txt', header=None, sep='\t')
ds.columns = ['users', 'songs', 'count']
```

```python
users = pd.read_csv('kaggle_users.csv', header=None)
users = users.reset_index()
users.columns = ['u_index', 'users']
```

```python
songs = pd.read_csv('kaggle_songs.csv', sep = ' ', header=None)
songs.columns = ['songs', 's_index']
```

```python
_users = pd.merge(ds, users, left_on='users', right_on='users', how='inner')
```

```python
users_songs = pd.merge(_users, songs, left_on='songs', right_on='songs', how='inner')
```

```python
users_songs = users_songs[['u_index', 's_index', 'count']]
```

```python
users_songs = users_songs.set_index(['u_index', 's_index'])
```

```python
rows = users_songs.index.get_level_values('u_index').values
columns = users_songs.index.get_level_values('s_index').values
data = users_songs['count'].values

count_mat = coo_matrix((data, (rows, columns)), shape=(users.shape[0]+1, songs.shape[0]+1))
```

```python
count_mat.shape
```

```python
def plot_coo_matrix(m):
    if not isinstance(m, coo_matrix):
        m = coo_matrix(m)
    fig = plt.figure(figsize=(20,20))
    ax = fig.add_subplot(111, axisbg='black')
    ax.plot(m.col, m.row, 's', color='white', ms=1)
    ax.set_xlim(0, m.shape[1])
    ax.set_ylim(0, m.shape[0])
    ax.set_aspect('equal')
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.invert_yaxis()
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])
    return ax
```

```python
import matplotlib.pyplot as plt
```

```python
ax = plot_coo_matrix(count_mat)
ax.figure.show()
```
